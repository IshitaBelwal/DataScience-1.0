```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[12,3,42]
y=[1,34,55]
x2=[1,23,30]
y2=[15,65,7]

plt.plot(x,y,label='first',linewidth=5)
plt.plot(x2,y2,label='second')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[12,3,42]
y=[1,34,55]
x2=[1,23,30]
y2=[15,65,7]

plt.scatter(x,y,label='first',linewidth=5)
plt.scatter(x2,y2,label='second')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[12,3,42]
y=[1,34,55]
x2=[1,23,30]
y2=[15,65,7]

plt.hist(x,y,label='first',linewidth=5)
plt.hist(x2,y2,label='second')
plt.legend()
plt.show()
```
